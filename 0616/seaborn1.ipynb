```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.countplot(data=tips, x='sex', hue='size') #x축에 column으로 나누고, 
#y축에 데이터의 개수를 센다. hue에 어떤 컬럼을 적으면 그 컬럼을 기준으로 데이터를 쪼갠다. 
```

```python
sns.barplot(data=tips, x='sex',y='tip', hue='size') 
```

```python
countplot과 차이는 barplot에는 y축을 설정하게 되어있다. 
숫자가 들어있는 column을 y축으로 설정한다.
countplot에는 y축을 default로 데이터 분석하는 
숫자에 소숫점이나 정수가 들어간다 하면 y축에 그 column을 넣고 barplot 
분석하는 것에 숫자가 필요 없으면 countplot 
label로 생각하면 된다. 

```
